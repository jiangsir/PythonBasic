```python
import tkinter as tk
from tkinter import messagebox

# 定義棋盤大小和格子大小
BOARD_SIZE = 15
GRID_SIZE = 40

# 定義棋盤狀態
EMPTY = 0
BLACK = 1
WHITE = 2

class GomokuGame:
    def __init__(self):
        self.board = [[EMPTY] * BOARD_SIZE for _ in range(BOARD_SIZE)]
        self.current_player = BLACK

    def make_move(self, row, col):
        if self.board[row][col] == EMPTY:
            self.board[row][col] = self.current_player
            if self.check_winner(row, col):
                winner = "Black" if self.current_player == BLACK else "White"
                messagebox.showinfo("Game Over", f"{winner} wins!")
                self.reset_game()
            else:
                self.switch_player()

    def check_winner(self, row, col):
        directions = [(1, 0), (0, 1), (1, 1), (-1, 1)]  # 檢查垂直、水平、斜線方向
        for d in directions:
            count = 1
            r, c = row, col
            dr, dc = d
            while count < 5:
                r += dr
                c += dc
                if 0 <= r < BOARD_SIZE and 0 <= c < BOARD_SIZE and self.board[r][c] == self.current_player:
                    count += 1
                else:
                    break
            r, c = row, col
            dr, dc = -d
            while count < 5:
                r += dr
                c += dc
                if 0 <= r < BOARD_SIZE and 0 <= c < BOARD_SIZE and self.board[r][c] == self.current_player:
                    count += 1
                else:
                    break
            if count >= 5:
                return True
        return False

    def switch_player(self):
        self.current_player = WHITE if self.current_player == BLACK else BLACK

    def reset_game(self):
        self.board = [[EMPTY] * BOARD_SIZE for _ in range(BOARD_SIZE)]
        self.current_player = BLACK

class GomokuGUI:
    def __init__(self, game):
        self.game = game

        self.root = tk.Tk()
        self.root.title("Gomoku")
        self.canvas = tk.Canvas(self.root, width=GRID_SIZE*BOARD_SIZE, height=GRID_SIZE*BOARD_SIZE)
        self.canvas.bind("<Button-1>", self.on_click)
        self.canvas.pack()

        self.draw_board()

        self.root.mainloop()

    def draw_board(self):
        for row in range(BOARD_SIZE):
            for col in range(BOARD_SIZE):
                x1 = col * GRID_SIZE
                y1 = row * GRID_SIZE
                x2 = x1 + GRID_SIZE
                y2 = y1 + GRID_SIZE

                if self.game.board[row][col] == BLACK:
                    self.canvas.create_oval(x1, y1, x2, y2, fill="black")
                elif self.game.board[row][col] == WHITE:
                    self.canvas.create_oval(x1, y1, x2, y2, fill="white")

    def on_click(self, event):
        row = event.y // GRID_SIZE
        col = event.x // GRID_SIZE
        self.game.make_move(row, col)
        self.canvas.delete("all")
        self.draw_board()

if __name__ == "__main__":
    game = GomokuGame()
    gui = GomokuGUI(game)
```
