### read()    read(size)   readline()   readlines()

　　　　之前的例子已經接觸到了 read() 函數，該函數會會一次性讀取文檔的全部內容，如果能確保文檔的大小，自然可以。但若文檔過大，內存就爆了，所以，可以反覆調用read(size)方法，每次最多讀取size個字節的內容；也可調用 readline() 每次讀取一行內容；而調用readlines()可以一次讀取所有內容並按行返回list。總之，根據需求來。僅以 txt 文檔為例，其他的文檔讀取需要特殊處理；另外，文檔的格式編碼方式也需要注意；這邊僅介紹讀取方法，其他的會出專題來學習。

## 利用 with open 讀取檔案,並自動關閉

```python
basepath = 'C:\\tmp\\'

with open(basepath+"Python生成測試檔.txt", encoding="UTF-8") as fp:
    print(fp.readlines())
    print('每一行最後都會附加一個 \\n 符號, 代表換行')
    
with open(basepath+"Python生成測試檔.txt", encoding="UTF-8") as fp:
    for line in iter(fp.readline, ""):
        print(line)
```

```python

basepath = 'D:\\'
def create(n):
    for i in range(n):
        f = open(basepath+'DESC_{:04d}.txt'.format(i), 'w', encoding = 'UTF-8')    
        # 也可使用指定路徑等方式，如： C:\A.txt
        f.write("測試文件的內容！！！")
        f.close()

create(10)
```

# 檢查 path 是否存在, 不存在就建立


```python
import os

path = os.path.join('C:\\', 'Users', 'user', 'Desktop')
dirname = '不存在的資料夾名稱' 
path = os.path.join(path, dirname)
print(path, os.path.dirname(path))
if not os.path.exists(path):
    print('os.makedir:', path)
    os.makedirs(path)
else:
    print('存在:', path)
```
